```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import a csv file
memory_cvlc_h264_trial1=pd.read_csv('raw_data/memory_cvlch264_trial1')
memory_cvlc_h264_trial2=pd.read_csv('raw_data/memory_cvlch264_trial2')
memory_cvlc_h264_trial3=pd.read_csv('raw_data/memory_cvlch264_trial3')
memory_cvlc_h264_trial4=pd.read_csv('raw_data/memory_cvlch264_trial4')
memory_cvlc_h264_trial5=pd.read_csv('raw_data/memory_cvlch264_trial5')
memory_cvlc_h264_trial6=pd.read_csv('raw_data/memory_cvlch264_trial6')
memory_cvlc_h264_trial7=pd.read_csv('raw_data/memory_cvlch264_trial7')
#concat all csv files
memory_cvlc_h264=pd.concat([memory_cvlc_h264_trial1,memory_cvlc_h264_trial2,memory_cvlc_h264_trial3,memory_cvlc_h264_trial4,memory_cvlc_h264_trial5,memory_cvlc_h264_trial6,memory_cvlc_h264_trial7])
#dataframe to retrieve column value
df1=pd.DataFrame(memory_cvlc_h264,columns=['value'])
#convert Bytes to Megabytes
df1['value']=df1['value']/(1000**2)
value1=df1['value']
#plot boxplot
plt.boxplot(value1)
plt.title('Memory_cvlc_h264')
plt.ylabel('value_in_MB')
plt.show()
value1
```
