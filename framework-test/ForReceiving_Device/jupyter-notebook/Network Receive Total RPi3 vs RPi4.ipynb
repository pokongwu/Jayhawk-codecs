```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import a csv file
network_receive_cvlc_h264_trial1=pd.read_csv('rp3_data/network_rx_cvlc-h264_trial1')
network_receive_cvlc_h264_trial2=pd.read_csv('rp3_data/network_rx_cvlc-h264_trial2')
network_receive_cvlc_h264_trial3=pd.read_csv('rp3_data/network_rx_cvlc-h264_trial3')
network_receive_cvlc_h264_trial4=pd.read_csv('rp3_data/network_rx_cvlc-h264_trial4')
network_receive_cvlc_h264_trial5=pd.read_csv('rp3_data/network_rx_cvlc-h264_trial5')
network_receive_cvlc_h264_trial6=pd.read_csv('rp3_data/network_rx_cvlc-h264_trial6')
network_receive_cvlc_h264_trial7=pd.read_csv('rp3_data/network_rx_cvlc-h264_trial7')
#concat all csv files
network_receive_cvlc_h264=pd.concat([network_receive_cvlc_h264_trial1,network_receive_cvlc_h264_trial2,network_receive_cvlc_h264_trial3,network_receive_cvlc_h264_trial4,network_receive_cvlc_h264_trial5,network_receive_cvlc_h264_trial6,network_receive_cvlc_h264_trial7])
#dataframe to retrieve column value
df1=pd.DataFrame(network_receive_cvlc_h264,columns=['value'])
#convert Bytes to Megabytes
df1['value']=df1['value']/(125000)
value1=df1['value']
#plot boxplot
plt.boxplot(value1)
plt.title('RPi3_Network_Receive_CVLC_H.264')
plt.ylabel('Value_in_Mbps')
plt.xticks([1], ['RPi3'])
plt.savefig("RPi3_Network_Receive_CVLC_H.264.png")
plt.show()
value1
```
