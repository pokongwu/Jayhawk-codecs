```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import a csv file
network_receive_cvlc_h264=pd.read_csv('net_rx_cvlch264_mod.csv')
#convert timestamp to datetime
network_receive_cvlc_h264['timestamp']=pd.to_datetime(network_receive_cvlc_h264['timestamp'],unit='s')
#convert Bytes to Megabytes
network_receive_cvlc_h264['value']=network_receive_cvlc_h264['value']/(1000**2)
#plot boxplot
ax1=network_receive_cvlc_h264.boxplot(by='timestamp',column=['value'],grid=False,figsize=(10,7))
plt.title('network_receive_cvlc_h264')
plt.ylabel('value')
plt.xticks(horizontalalignment='right',rotation=45)
#print result
result=[network_receive_cvlc_h264['timestamp'],network_receive_cvlc_h264['value']]
result
```
