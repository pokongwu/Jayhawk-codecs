```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import a csv file
network_transmit_cvlc_h264=pd.read_csv('net_tx_cvlch264_mod.csv')
#dataframe to retrieve column value
df1=pd.DataFrame(network_transmit_cvlc_h264,columns=['value'])
#convert Bytes to Megabytes
df1['value']=df1['value']/(1000**2)
value1=df1['value']
#plot boxplot
plt.boxplot(value1)
plt.title('network_transmit_cvlc_h264')
plt.ylabel('value_in_MB')
plt.show()
value1
```
