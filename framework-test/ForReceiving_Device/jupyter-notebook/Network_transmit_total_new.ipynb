```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import a csv file
network_transmit_cvlc_h264=pd.read_csv('net_tx_cvlch264_mod.csv')
#convert timestamp to datetime
network_transmit_cvlc_h264['timestamp']=pd.to_datetime(network_transmit_cvlc_h264['timestamp'],unit='s')
#convert Bytes to Megabytes
network_transmit_cvlc_h264['value']=network_transmit_cvlc_h264['value']/(1000**2)
#plot boxplot
ax1=network_transmit_cvlc_h264.boxplot(by='timestamp',column=['value'],grid=False,figsize=(10,7))
plt.title('network_transmit_cvlc_h264')
plt.ylabel('value')
plt.xticks(horizontalalignment='right',rotation=45)
#print result
result=[network_transmit_cvlc_h264['timestamp'],network_transmit_cvlc_h264['value']]
result
```
