```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import a csv file
cpu_cvlc_h264_trial1=pd.read_csv('raw_data/cpu_cvlch264_trial1')
cpu_cvlc_h264_trial2=pd.read_csv('raw_data/cpu_cvlch264_trial2')
cpu_cvlc_h264_trial3=pd.read_csv('raw_data/cpu_cvlch264_trial3')
cpu_cvlc_h264_trial4=pd.read_csv('raw_data/cpu_cvlch264_trial4')
cpu_cvlc_h264_trial5=pd.read_csv('raw_data/cpu_cvlch264_trial5')
cpu_cvlc_h264_trial6=pd.read_csv('raw_data/cpu_cvlch264_trial6')
cpu_cvlc_h264_trial7=pd.read_csv('raw_data/cpu_cvlch264_trial7')
#concat all csv files
cpu_cvlc_h264=pd.concat([cpu_cvlc_h264_trial1,cpu_cvlc_h264_trial2,cpu_cvlc_h264_trial3,cpu_cvlc_h264_trial4,cpu_cvlc_h264_trial5,cpu_cvlc_h264_trial6,cpu_cvlc_h264_trial7],axis=0)
#convert timestamp to datetime
cpu_cvlc_h264['timestamp']=pd.to_datetime(cpu_cvlc_h264['timestamp'],unit='s')
#convert Bytes to Megabytes
cpu_cvlc_h264['value']=cpu_cvlc_h264['value']/(1000**2)
#plot boxplot
ax1=cpu_cvlc_h264.boxplot(by='timestamp',column='value',grid=False,figsize=(10,7))
plt.title('cpu_cvlc_h264')
plt.ylabel('value')
plt.xticks(horizontalalignment='right',rotation=45)
#print result
result=[cpu_cvlc_h264['timestamp'],cpu_cvlc_h264['value']]
result
```
